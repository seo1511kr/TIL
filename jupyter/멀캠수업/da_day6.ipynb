```python
#데이터프레임 합성:merge:공통된 열 또는 인덱스를 기준으로 합침, concat
#merge:공통된 열(key) 또는 인덱스를 기준으로 합침
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
pd.merge(df1,df2) #how="inner"가 디폴트, 생략됨
#merge:데이터프레임 합치기, 합치는 방법이 다양하게 존재. 기본은 내부조인
#내부조인(inner join):두 개의 데이터프레임에 공통으로 존재하는 키를 중심으로 함침
```

```python
#외부조인(outer join):키(공통 컬럼) 값이 어느 한쪽에만 있어도 병합이 됨
pd.merge(df1,df2, how="outer")
```

```python
pd.merge(df1, df2, how="left")
#df1(left)을 기준으로 df2와 merge
#df1에 있는 모든 데이터는 merge가 됨.
#df2에 있는 데이터 중에 df1에 존재하지 않는 데이터는 merge가 안됨
```

```python
pd.merge(df1, df2, how="right")
#df2(right)을 기준으로 df1와 merge
#df2에 있는 모든 데이터는 merge가 됨.
#df1에 있는 데이터 중에 df2에 존재하지 않는 데이터는 merge가 안됨
```

```python
np.random.seed(0)
df2 = pd.DataFrame({
    'key1': ['A', 'A', 'B', 'B', 'A'],
    'key2': ['one', 'two', 'one', 'two', 'one'],
    'data1': [1, 2, 3, 4, 5],
    'data2': [10, 20, 30, 40, 50]
})
df2
```

```python
#그룹 A, 그룹 B로 구분
groups=df2.groupby(df2.key1)
groups
```

```python
groups.sum()
```

```python
group1=df2.data1.groupby(df2.key1)
group1
#key1을 기준으로 data1열의 데이터가 그루핑 됨
```

```python
#그룹화 연산시 사용자 정의함수(최대값과 최소값의 차이를 계산하는 함수)를 호출하여 수행
```

```python
def diff(x): #x에는 종별 그룹이 전달됨
    #return x.max()-x.min() #그룹 단위로 컬럼별 max, min 연산 수행
    return x.max()/x.min()
```

```python
iris.groupby(iris.species).agg(diff)
#iris의 종별 그룹화 -> 그룹 단위로 diff함수를 적용(agg:aggregate)
```
