# 그룹화한 데이터의 평균 구하기

```python
df.head()
```

```python
df.groupby(by='country').groups
```

```python
df.groupby(by='country').mean()
```

```python
def minmax(x):
    return x.min()+x.max()
df.groupby(by='country').lifeExp.agg([minmax,min])
```

```python
df.groupby(by='country').apply(lambda x:x.sort_values(by='pop',ascending=True))
```
