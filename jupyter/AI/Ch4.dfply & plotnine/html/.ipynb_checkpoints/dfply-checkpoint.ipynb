# 3. Data Manipulation

### 3.1 mutate()

### 3.2 Grouping

ungroup()

### 3.3 Summarization

summarize_each([np.mean, np.var], 'price', 'depth')
summarize(cut_n = n(X.price))
summarize(price_ndistinct = n_distinct(X.price))
summarize(price_mean=mean(X.price),
          price_iqr=IQR(X.price),
          price_min=colmin(X.price),
          price_max=colmax(X.price),
          price_median=median(X.price),
          price_var=var(X.price),
          price_sd=sd(X.price)))
