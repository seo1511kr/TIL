```python
driver=webdriver.Chrome("c:/scrap/chromedriver.exe")
url_main="https://news.naver.com/main/list.nhn?mode=LS2D&sid2=260&mid=shm&sid1=101&date=202101"

# url="https://news.naver.com/main/list.nhn?mode=LS2D&sid2=260&sid1=101&mid=shm&date=20210126&page="
import re
title_Jan=[]
for k in range(26):
    if k < 9:
        url_date=url_main+"0"+str(k+1)
    else:url_date=url_main+str(k+1)
    driver.get(url_date)
    num_pages=len(driver.find_elements_by_css_selector("#main_content > div.paging > a"))
    title_pages=[]
    for i in range(num_pages):
        url_page=url_date+"&page="+str(i+1)
        driver.get(url_page)
        time.sleep(2)
        html=driver.page_source
        soup=BeautifulSoup(html,"html.parser")
        title_li=soup.select("dl > dt:nth-child(2)")
        num=len(title_li)
        for j in range(num):
            title=title_li[j].text
            if re.search("(상승)|(역대)|(껑충)|(과열)|(높)|(날)|(올랐)|(폭등)|(오른)",title):
                title_pages.append(title)
    title_Jan.append([str(i+1)+"일:",title_pages])
```

```python
print(title_Jan)
```

```python
for i in title_Jan:
    for j in range(len(i[1])):
        i[1][j]=i[1][j].strip("\n").strip("\t").strip("\n")
```
